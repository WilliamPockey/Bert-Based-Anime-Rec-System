## 分词

```python
import pandas as pd
import jieba
import re
import string
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rc("font", family='Microsoft YaHei')

# 读取CSV文件
df = pd.read_csv('anime.csv', encoding='gbk')
# print(df)

# 去除虚词、标点和空白
def remove_stopwords_and_punctuations(text):
    stopwords = set(['这种','着','还是','因为','级','得','有点','导演','接受','a','第五','其他','样子','看着','部','12','卷','前面','日','那个','事','不明','懂','认为','本文','成为','每','4','低分','两部','地区','一天','嘞','图','1','一季','想要','下巴','尖','理解','重现','不了','の','老妈','父亲','九年','挺','这是','内容','补上','长评','总之','咱','这里','作者','ー','を','为什么','9年','第四季','依然','粮','满屏','题材','9','结果','之后','将','过','以后','分','等','来看','评论','为了','组','大','勇','一','一集','不同','某些','剧中','布','一下','站','对于','对方','一起','ED','为了','版','TV','因此','某','不行','5','时而','您','以及','篇','不会','一点','无法','评分','却','年','却','只有','全部','知道','不能','首先','并','为','一样','里面','专栏','来源于','九秒','然后','的话','弹幕','帅帅','刻画','话','时','非常','方面','根线','突然','啊','部分','前期','吗','已经','这么','如果','似的','跟','一次','通过','所以','应该','那','之一','一种','里','其实','这','问题','第三集','想','反正','以来','优点','缺点','比较','有些','校长','这样','太','现在','完全','第二季','第三季','挑限','进度','很长','子','从','较为','感','小','第一集','贴','这番','微博上','虽然','们','或者','不过','相比','角色','主角','开始','这样','老子','b','截图','监督','再','甚至','观众','才','一些','s','伊','露露','像','在','可以','武本','评价','','第一季','第二季','因为','作为','而且','属于','看到','下','一句','一直','要','番','来说','表现','人物','动漫','只能','还有','集','那么','比','可能','什么','真正','sb','剧','剧本','确实','足够','出','加上','打','当时','带来','点','去','后','前','怎么','那种','动画','简直','的', '了', '和', '是', '在', '我','我们','你','你们','它','它们','她','她们','他','他们', '有', '不', '这', '个', '也', '就', '很', '还', '就是', '而', '及', '或', '与', '之', '但', '之类','看','听','到','作品','会','都','喜欢','感觉','说','人','最','很多','上','做','让','被','把','没有','完','给','好','时候','一个','自己','更','把','吧','这个','最后','后面','真的','觉得','这部','中','能','没','俺','所','一部','呢','但是','出来','来','多','啥','又','只是','每个','不是','对','特别'])
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    text = re.sub("\n", "", text)
    text = text.replace(" ", "")
    words = jieba.lcut(text)
    filtered_words = [word for word in words if word not in stopwords and word.strip() != '' and word != '\n']
    return ''.join(filtered_words)


# 分词并统计每行的最高词频
def get_highest_word_frequency(row):
    cleaned_comment = remove_stopwords_and_punctuations(row['Comments'])
    words = jieba.lcut(cleaned_comment)
    word_count = {}
    for word in words:
        word_count[word] = word_count.get(word, 0) + 1
    highest_frequency_words = sorted(word_count.items(), key=lambda x: x[1], reverse=True)[:10]
    return highest_frequency_words



# 添加新列，存储每行的最高词频
df['highest_frequency_words'] = df.apply(get_highest_word_frequency, axis=1)
df = df.drop(columns=['Comments'])
df = df.drop(columns=df.columns[0])

split_columns = df['Tags'].str.split('/', expand=True)
# 将拆分后的列添加到 DataFrame
for index, row in split_columns.iterrows():
    for value in row:
        if value is not None:
            if f'{value}' not in df.columns:
                df[f'{value}'] = 0
            df[f'{value}'][index] = 1
    
    
df = df.drop(columns=['Tags'])
tags = [tag for sublist in df['highest_frequency_words'] for tag, value in sublist]
# print(tags)
values = [value for sublist in df['highest_frequency_words'] for tag, value in sublist]

for i,col in enumerate(tags):
    # df[f'L{col}'] = values[i]
    if f'{col}' not in df.columns:
                df[f'{col}'] = 0
    # print(int(i/9))
    #有多少标签
    print(int(i/10),f'{col}')
    df.loc[int(i/10), f'{col}'] = 1



# 绘制柱状图
for i, row in df.iterrows():
    if not row['highest_frequency_words'] == []:
        plt.figure(figsize=(8, 6))
        print(i)
        print(row)
        print(row['highest_frequency_words'])
        words, frequencies = zip(*row['highest_frequency_words'])
        plt.bar(words, frequencies)
        plt.xlabel(row['Title'])
        plt.ylabel('标签频率')
        plt.title(f'第{i+1}部番剧的最高标签频率')
        plt.xticks(rotation=45)
        plt.show()

df = df.drop(columns=['highest_frequency_words'])
df.to_csv('train.csv',index=False)
```
